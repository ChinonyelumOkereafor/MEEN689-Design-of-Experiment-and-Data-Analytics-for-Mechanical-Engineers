a. Identify the type of t-test to be used and explain.

#Answers
Type of T-test: **Paired samples t-test using two-tailed** (also known as a dependent samples t-test or paired t-test)

**Explaination**

**Paired samples t-test:** Since each car tests both tire brands, each pair of mileage values (Brand "X" and Brand "Y") is dependent. The paired t-test is suitable here, as it evaluates whether the average difference in wear (mileage) between the brands is significantly different from zero.

**Two-tailed test:** From the question,we're checking if there’s a **difference** in wear rate between two tire brands, without specifying direction. Two-tailed test is appropriate because the difference could favor either brand. At 95% confidence, we would examine both distribution tails for statistical significance in either direction.

b. Calculate the degrees of freedom, t-value from the table, and t-value from the data.

```python
#Data from the Datasheet Q1
brand_x = df1['Brand “X”']
brand_y = df1['Brand “Y”']

# Calculate the differences
differences = brand_x - brand_y

# Degrees of freedom (n - 1 for paired samples t-test)
n = len(differences)
degrees_of_freedom = n - 1

# Calculate t-value from the data
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)
t_value_data = mean_diff / (std_diff / np.sqrt(n))

# Find the critical t-value from the table for 95% confidence level (two-tailed)
t_value_table = stats.t.ppf(1 - 0.05/2, degrees_of_freedom)


#Displaying results
print("================================================================")
print("Results for Two-tailed Paired samples t-test Test:")
print("----------------------------------------------------------------")
print("Degrees of Freedom:", degrees_of_freedom)
print("Mean Difference:", mean_diff)
print("Standard Deviation of Differences:", std_diff)
print("t-value from the data/sample:", t_value_data)
print("t-value from the table (Critical t-value):", t_value_table)
print("===============================================================")
```

a. Assuming pooled standard deviation of 3 s, run the power analysis to determine the sample size needed for this study for an 80% power.

```python
# Parameters for power analysis
Mean_difference = 4  # Given Mean difference (4 seconds)
pooled_std_dev = 3  # Given Pooled standard deviation (3 seconds)
effect_size = Mean_difference/pooled_std_dev
alpha = 0.05  # Significance level
power = 0.80  # Desired power level

# Perform power analysis to calculate sample size needed for each group
analysis = TTestIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, power=power, alpha=alpha, alternative='two-sided')

print("================================================================")
print("Power Analysis Results:")
print("----------------------------------------------------------------")
print("Sample Size Needed for Each Group:", sample_size)
print("Approximated Sample Size:", np.ceil(sample_size))
print("===============================================================")
```

**Comments**

The power analysis indicates that approximately 10 participants per group are needed to achieve 80% power for detecting a mean difference of 4 seconds with a pooled standard deviation of 3 seconds, at a 5% significance level.

From the question, this is a two-tailed test. Therefore, using a two-sided test would be less efficient in this context requiring a larger sample size to maintain the same power and introducing the possibility of detecting a significant result in an irrelevant direction (i.e., worse performance by the trained group).

```python
# Calculating variances of the trained and untrained groups
variance_trained = np.var(df2['Trained'], ddof=1)  # Sample variance
variance_untrained = np.var(df2['Untrained'], ddof=1)  # Sample variance

print("================================================================")
print("Variance Results:")
print("----------------------------------------------------------------")
print(f"Variance of Trained Group: {variance_trained:.6f}")
print(f"Variance of Untrained Group: {variance_untrained:.6f}")
print("--->Therefore, the variances are equal")
print("===============================================================")
```

```python
# Data for trained and untrained groups
trained = df2['Trained']
untrained = df2['Untrained']

# Calculate t-critical value for a two-tailed test with degrees of freedom based on sample size
n_trained = len(trained)
n_untrained = len(untrained)

#Mean difference
mean_diff = np.mean(untrained) - np.mean(trained)

degrees_of_freedom = n_trained + n_untrained - 2

# t-critical for 95% confidence level (two-tailed)
t_critical = stats.t.ppf(1 - 0.05/2, degrees_of_freedom)

# Perform an independent samples t-test
t_statistic, p_value = stats.ttest_ind(trained, untrained, equal_var=True)

print("================================================================")
print("Independent Samples t-test Results:")
print("----------------------------------------------------------------")
print("Mean Difference:", mean_diff)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
print("Critical t-value:", t_critical)
print("===============================================================")
```

**Conclusion**

Since the absolute t-statistic (2.636) exceeds the t-critical value (2.120), **we reject the null hypothesis.** This confirms that the training course has a statistically significant effect on performance time.

d. Is it necessary that post-hoc power (Part c) to be the same as priori power (Part a)? Do literature search and explain.

#Response to Question 1d
No, post-hoc power (calculated after the experiment based on observed effect size) is not necessarily the same as priori power (calculated before the experiment based on an expected effect size).



**Literature Search**

Research highlights that post-hoc power has limited practical value, as it relies heavily on the observed effect size, which can vary due to sampling. According to Hoenig and Heisey (2001), their study suggested that post-hoc power does not reliably assess a study's design or sample size adequacy. Instead, emphasis should be on a priori power to ensure a robust study design.

In summary, post-hoc and a priori power serve distinct purposes and need not align, as they are influenced by different factors in study planning.

.

**Explanation**

**Priori power** is calculated before a study begins to estimate the sample size needed to detect a desired effect, usually aiming for a power level like 80%. This ensures there is a high chance of detecting a true effect if one exists, guiding study planning.

**Post-hoc power** is calculated after data collection based on the actual effect size observed. It shows the study’s power to detect the effect that occurred not the initially expected effect.

Since the observed effect size often varies from the anticipated effect due to sampling variability, post-hoc power can differ from priori power thereby leading it to be higher or lower than planned.


**References**

Hoenig, J. M., & Heisey, D. M. (2001). The Abuse of Power: The Pervasive Fallacy of Power Calculations for Data Analysis. The American Statistician, 55(1), 19–24.

a. Calculate expected frequencies and determine χ2obs. Show your calculations.

```python
# Given data
data = {
    "Process Line": df3['Process Line'],
    "Success": df3['Success'],
    "Failure": df3['Failure']
}
df = pd.DataFrame(data)

# Total successes, failures, and grand total
total_success = df["Success"].sum()
total_failure = df["Failure"].sum()
grand_total = total_success + total_failure

# Calculating expected frequencies for each cell in a 5x2 table
df["Expected Success"] = ((df["Success"] + df["Failure"]) * total_success) / grand_total
df["Expected Failure"] = ((df["Success"] + df["Failure"]) * total_failure) / grand_total

# Calculating chi-square observed value
chi_square_obs = (((df["Success"] - df["Expected Success"]) ** 2) / df["Expected Success"]).sum() + \
                 (((df["Failure"] - df["Expected Failure"]) ** 2) / df["Expected Failure"]).sum()


print("=================================================================================")
print("Calculated Expected Frequencies:")
print("---------------------------------------------------------------------------------")
print(df[['Process Line', 'Success', 'Failure', 'Expected Success', 'Expected Failure']])
print("---------------------------------------------------------------------------------")
print("Total Success:", total_success)
print("Total Failure:", total_failure)
print("Grand Total:", grand_total)
print("=================================================================================")
print("Chi-square Observed Value:", chi_square_obs)
print("=================================================================================")
```

b. Determine the degrees of freedom and the X2crit

```python
# Degrees of freedom for chi-square test in a contingency table
# (rows - 1) * (columns - 1)
rows = 5  # process lines
columns = 2  # success and failure
degrees_of_freedom = (rows - 1) * (columns - 1)

# Chi-square critical value for 95% confidence level
alpha = 0.05
chi_square_critical = stats.chi2.ppf(1 - alpha, degrees_of_freedom)

print("================================================================")
print("Degrees of Freedom:", degrees_of_freedom)
print("Chi-square Critical Value:", chi_square_critical)
print("===============================================================")
```
